```python
# Function for outliers
def detect_outliers_iqr(df, columns):
    outlier_summary = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        outlier_summary[col] = len(outliers)
    return pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier Count'])
```

```python
# List all numeric columns
numeric_cols = test.select_dtypes(include=['float64']).columns
numeric_cols
```

```python
# Detect outliers
outliers_df = detect_outliers_iqr(test, numeric_cols)
outliers_df.sort_values(by='Outlier Count', ascending=False)
```

```python
# Select only columns with at least 1 outlier
cols_with_outliers = outliers_df[outliers_df['Outlier Count'] > 0].index.tolist()
cols_with_outliers

plt.figure(figsize=(16, 12))
for i, col in enumerate(cols_with_outliers):
    plt.subplot((len(cols_with_outliers) + 2) // 3, 3, i + 1)
    sns.boxplot(x=test[col])
    plt.title(col)
    plt.tight_layout()

plt.suptitle("Boxplots for Outlier Detection", fontsize=18, y=1.02)
plt.show()
```

# OUTLIER ANALYZIS
- 'phone' and 'tabswitched' have narrow IQRs with outliers near 0, suggesting rare instances of phone use or tab switching during tests.
- 'proctoring_analysis' shows a wide IQR (100-500) with outliers above 600, indicating varying proctoring issues, with some extreme cases.
- 'candidate_total_score' has a tight IQR (20-30), with outliers near 0, suggesting most scores cluster around this range, with a few low performers.
- 'data_interpretation_candidate_score' and 'tech_and_computing_fundamentals_candidate_score' have narrow IQRs (2-4 and 1-5 respectively), with minimal outliers, indicating consistent performance.
- 'general_reasoning_candidate_score' has a tight IQR (6-8), with one outlier near 0, showing uniform reasoning skills with rare exceptions.
- 'token_paid' has a wide spread with multiple outliers (1-7 range), suggesting inconsistent token payments or data recording issues.
