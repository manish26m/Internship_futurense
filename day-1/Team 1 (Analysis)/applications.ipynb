```python
total_rows = len(applications)

# total number of nulls in each column
null_counts = applications.isnull().sum()

# total percentage of null values in each column
null_percentages = (null_counts / total_rows * 100).round(2)

null_summary = pd.DataFrame({'Null Count': null_counts,'Null Percentage (%)': null_percentages})

print("Null Value Summary:")
print(null_summary)
```

Overall, most fields are complete (0% null), but high null percentages in post-graduation and token-related fields suggest these may not apply to all candidates.

```python
# Detect outliers using IQR method
def detect_outliers_iqr(df, columns):
    outlier_summary = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        outlier_summary[col] = len(outliers)
    return pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier Count'])
```

```python
# Get numeric columns (float64 and int64)
numeric_cols = applications.select_dtypes(include=['float64', 'int64']).columns
numeric_cols
```

```python
outlier_counts = detect_outliers_iqr(applications, numeric_cols)

# Display the counts sorted by number of outliers descending
outlier_counts_sorted = outlier_counts.sort_values(by='Outlier Count', ascending=False)
outlier_counts_sorted
```

```python
# Filter numeric columns to keep only those with at least one non-null value
valid_numeric_cols = [col for col in numeric_cols if applications[col].dropna().shape[0] > 0]

plt.figure(figsize=(16, len(valid_numeric_cols) * 3))

for i, col in enumerate(valid_numeric_cols, 1):
    plt.subplot(len(valid_numeric_cols), 1, i)
    sns.boxplot(x=applications[col], color='skyblue')
    plt.title(f'Boxplot for {col}')
    plt.tight_layout()

plt.suptitle('Boxplots of Numeric Columns', fontsize=16, y=1.02)
plt.show()
```

# OUTLIER ANALYZIS
- 'batch' and 'token' boxplots show tight IQRs around 1.0 with no outliers, indicating consistent values.
- 'phone' has a narrow IQR (0.7-0.9) with minimal variation and no significant outliers.
- 'permanent_pincode' IQR spans 200000-600000, suggesting a wide range of geographic data with some outliers.
- 'x_passing_year' and 'xii_year' IQRs center around 2005-2015, with outliers before 1990, indicating older educational records.
- 'bachelor_passing_year' and 'post_graduation_passing_year' IQRs are around 2010-2020, with outliers pre-2000, showing recent graduations with some exceptions.
- 'bachelor_passing_percentage' and 'post_graduation_percentage' IQRs are 60-80%, with outliers near 0, suggesting mostly consistent performance with rare low scores.
- 'year_of_experience' has a narrow IQR (5-20) with an outlier near 100, possibly a data error.
- 'test' IQR is tight around 1.0 with no outliers, indicating uniform test data.

```python
# Email → Phone
email_to_phone = applications.groupby('email')['phone'].nunique()
emails_multi_phones = (email_to_phone > 1).sum()
print(f"Emails linked to multiple phones: {emails_multi_phones}")
```

```python
# Phone → Email
phone_to_email = applications.groupby('phone')['email'].nunique()
phones_multi_emails = (phone_to_email > 1).sum()
print(f"Phones linked to multiple emails: {phones_multi_emails}")
```

```python
# Candidate ID → Phone/Email
candidate_to_email = applications.groupby('candidate_id')['email'].nunique()
candidate_to_phone = applications.groupby('candidate_id')['phone'].nunique()
cids_multi_emails = (candidate_to_email > 1).sum()
cids_multi_phones = (candidate_to_phone > 1).sum()

print(f"Candidate IDs linked to multiple emails: {cids_multi_emails}")
print(f"Candidate IDs linked to multiple phones: {cids_multi_phones}")
```
